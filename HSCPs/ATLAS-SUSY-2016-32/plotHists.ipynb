```python
import matplotlib.pyplot as plt
import os
import numpy as np
import seaborn as sns
from scipy.special import kn,zetac
sns.set() #Set style
sns.set_style('ticks',{'font.family':'serif', 'font.serif':'Times New Roman'})
sns.set_context('paper', font_scale=1.8)
sns.set_palette(sns.color_palette("Paired"))
```

```python
recHist = np.genfromtxt('recHist.dat',names=True,delimiter=',')
recData = np.array([[eta,beta,w] for eta,beta,w in recHist])
```

```python
plt.hist2d(recData[:,0],recData[:,1],bins=10,cmap='Blues',weights=recData[:,2])
plt.xlabel(r'$|\eta|$')
plt.ylabel(r'$\beta$')
plt.xlim(recData[:,0].min(),recData[:,0].max())
plt.ylim(recData[:,1].min(),recData[:,1].max())
plt.show()
```

```python
massHist = np.genfromtxt('massHist.dat',names=True,delimiter=',')
massData = np.array([[mdEdx,mToF,w] for mdEdx,mToF,w in massHist])
```

```python
plt.hist2d(massData[:,0],massData[:,1],bins=20,cmap='Blues',weights=massData[:,2])
plt.xlabel(r'$m_{dE/dx}$ (GeV)')
plt.ylabel(r'$m_{ToF}$ (GeV)')
plt.xlim(massData[:,0].min(),massData[:,0].max())
plt.ylim(massData[:,1].min(),massData[:,1].max())
plt.show()
```
